```python
import pandas as pd

input_path  = "/mnt/cephfs-mount/chenchen/CygnssDataCsvLand_GWP_merged.csv"
output_path = "/mnt/cephfs-mount/chenchen/CygnssDataCsvLand_GWP_merged_10pct.csv"

# Read CSV
df = pd.read_csv(input_path)

# ----- Row counts -----
total_rows = len(df)
print("Total rows:", total_rows)

# Sample 10%
df_sample = df.sample(frac=0.10, random_state=42)

sample_rows = len(df_sample)
print("Sampled rows:", sample_rows)

# Percentage check
print(f"Sample fraction: {sample_rows / total_rows * 100:.2f}%")

# Save sample
df_sample.to_csv(output_path, index=False)
print("Sample saved to:", output_path)
```
