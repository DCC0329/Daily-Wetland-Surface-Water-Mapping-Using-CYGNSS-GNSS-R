```python
import os
import pandas as pd

# Root folder containing subfolders of CSVs
root = "/mnt/cephfs-mount/chenchen/CygnssDataCsvLand_GWP"

# Output file
output_path = "/mnt/cephfs-mount/chenchen/CygnssDataCsvLand_GWP_merged.csv"

all_dfs = []

# Walk through all subdirectories
for dirpath, _, filenames in os.walk(root):
    for fn in filenames:
        if fn.lower().endswith(".csv"):
            fpath = os.path.join(dirpath, fn)
            try:
                df = pd.read_csv(fpath)
                df["source_file"] = fn   # optional: track origin
                all_dfs.append(df)
            except Exception as e:
                print(f"‚ö†Ô∏è Failed to read: {fpath} ‚Äî {e}")

# Merge all CSVs together
if len(all_dfs) > 0:
    merged = pd.concat(all_dfs, ignore_index=True)
    merged.to_csv(output_path, index=False)
    print(f"‚úÖ Merged {len(all_dfs)} CSV files.")
    print(f"üìÑ Saved to: {output_path}")
else:
    print("‚ùå No CSV files found.")
```
