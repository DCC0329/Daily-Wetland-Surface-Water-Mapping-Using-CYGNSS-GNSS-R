```python
import pandas as pd

path = "/mnt/cephfs-mount/chenchen/CygnssDataCsvLand_GWP/2018_09/20180901.csv"

df = pd.read_csv(path)

# Ensure column exists
if "water_fraction" not in df.columns:
    raise Exception("Column 'water_fraction' not found")

vals = df["water_fraction"].dropna()

count_gt0 = (vals > 0).sum()
count_eq0 = (vals == 0).sum()

print("Total rows:", len(vals))
print("water_fraction > 0:", count_gt0)
print("water_fraction == 0:", count_eq0)
```
